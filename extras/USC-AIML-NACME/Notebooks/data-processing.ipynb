```python

import matplotlib.pyplot as plt
import numpy as np

# Which airport is the busiest
# (Explore the data programmatically/visually)
################################################################

# We decided to determine which airport had the most flights annually
# This can be done by using the destination column 'Dest' and departure column 'Origin'
# This is because all of the data is from a single year (2008)
# This will allow us to create counts for each airport and compare them in one chart
# We will also need to remove cancelled flights as they may skew data incorrectly

# Checking for rows we may need to eliminate due to null values:
print(f"Is there null values in the destination column: {df['Dest'].isnull().any()}")
print(f"Is there null values in the departure column: {df['Origin'].isnull().any()}")

# Discovering there are 304 unique destination airports (too many for a single bar chart)
# We will just display the top 10 values in our chart for visual simplicity
print(f"How many unique values are there in the destinations: {df['Dest'].unique().size}")
print(f"What is shown when flights are cancelled: {df['CancellationCode'].unique()}")

# removing the cancelled flights
df = df[df['CancellationCode'] == 'N']
print(f"This is the only code left after data processing: {df['CancellationCode'].unique()}")

# creating a series that contains the amount of flights to each destination in ascending order
departures_per_airport = df['Year'].groupby(df['Origin']).count()
arrivals_per_airport = df['Year'].groupby(df['Dest']).count()
flights_per_airport = departures_per_airport + arrivals_per_airport
flights_per_airport = flights_per_airport.sort_values()

#removing null values
flights_per_airport = flights_per_airport[~flights_per_airport.isnull()]

# pulling out the top 10 destinations with the most flights to prep for plot
top_10_airports = tuple(flights_per_airport.tail(10).index)
top_10_airport_flights = tuple(flights_per_airport.tail(10).values)

#Creating the plot and highlighting the busiest airport
question1_df = pd.DataFrame({
    'Top 10 Airports': top_10_airports,
    'Flight Counts': top_10_airport_flights,
})
x_coords = np.arange(len(question1_df))

colors = ['#0000FF' for _ in range(len(question1_df))]
colors[-1] = '#FF0000'

plt.bar(x_coords, question1_df['Flight Counts'], tick_label=question1_df['Top 10 Airports'], color=colors)
plt.xticks(rotation=90)
plt.ylabel('Flights Arrived at Airport')
plt.title('2008 Top 10 Busiest Airports')
plt.show()

################################################################
```
